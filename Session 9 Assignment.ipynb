```python
import pandas as pd

# Reading CSV file
baby_names = pd.read_csv('https://raw.githubusercontent.com/guipsamora/pandas_exercises/master/06_Stats/US_Baby_Names/US_Baby_Names_right.csv')
#Getting information of US_Baby_Names_right 
baby_names.info()
```

```python
#Display the first 10 entries
baby_names.head(10)
```

```python
# 1. Delete unnamed columns 
del baby_names['Unnamed: 0']
baby_names.head(10)
```

```python
# 2. Show the distribution of male and female 
baby_names['Gender'].value_counts()
```

```python
# 3. Show the top 5 most preferred names 
names=baby_names.groupby("Name").sum()  # Groupby Name and adding
names.sort_values("Count", ascending = 0).head() # sorting in acsending order
```

```python
# 4. What is the median name occurence in the dataset 
names[names.Count == names.Count.median()].head(10)
```

```python
# 5. Distribution of male and female born count by states 
genderbyState =baby_names.groupby(['Gender','State'])['Count'].count()
genderbyState
```
